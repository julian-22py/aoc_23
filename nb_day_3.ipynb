Part 1

```python
def get_neighbour_pos(line_id, start, end):
    pos_neighbours = []
    # line delta -1
    len = end - start + 1
    for dist in range(-1,len+1):
        pos_neighbours.append([line_id - 1, start + dist])

    # line delta 0
    for dist in [-1, len]:
        pos_neighbours.append([line_id, start + dist])

    # line delta 1
    for dist in range(-1,len+1):
        pos_neighbours.append([line_id + 1, start + dist])

    return pos_neighbours

get_neighbour_pos(4, 10, 12)
```

```python
def check_common_elelemts(list1, list2):
    return any([element in list2 for element in list1])

list1 = [1 , 2, 3]
list2 = [9, 2, 1]

check_common_elelemts(list1, list2)
```

```python
def get_common_elements(list1, list2):
    return [element for element in list1 if element in list2]

list1 = [[1 , 2, 3], [5, 6, 8]]
list2 = [[1 , 2, 3], [5, 7, 8]]

get_common_elements(list1, list2)
```

```python
# Find positions of special character

pos_special_char = []
special_char = []

for line_id, line in enumerate(lines):
    matches = re.finditer(r'[^\.\d]', line)

    for match in matches:
        special_char.append(match.group(0))
        #print(match.group(0), [line_id, match.span()[0]])
        pos_special_char.append([line_id, match.span()[0]])

print(
    pos_special_char[:5],
    set(special_char)
)
```

```python
valid_ids = []

for line_id, line in enumerate(lines):
    matches = re.finditer(r'\d+', line)

    for match in matches:
        #print(match.group(0), [line_id, match.span()[0], match.span()[1]-1])
        pos_neighbours = get_neighbour_pos(line_id, start=match.span()[0], end=match.span()[1]-1)
        if check_common_elelemts(pos_neighbours, pos_special_char):
            valid_ids.append(match.group(0))
            #print(match.group(0), [line_id, match.span()[0], match.span()[1]-1])
```

```python
result = sum([int(i) for i in valid_ids])
print(f"Answer of puzzle 3 part 1 is {result}!")
```

```python
numbers_with_star_neighbour= []
position_star_neighbour = []

for line_id, line in enumerate(lines):
    matches = re.finditer(r'\d+', line)

    for match in matches:
        #print(match.group(0), [line_id, match.span()[0], match.span()[1]-1])
        pos_neighbours = get_neighbour_pos(line_id, start=match.span()[0], end=match.span()[1]-1)

        #print(match, pos_neighbours)
        # Check if star are in the neighbourhood (one field away in every direction)
        if check_common_elelemts(pos_neighbours, pos_special_star):
            numbers_with_star_neighbour.append(match.group(0))

            position_of_star_in_neighbourhood = get_common_elements(pos_neighbours, pos_special_star)[0]

            #print(match.group(0), position_of_star_in_neighbourhood)
            position_star_neighbour.append(position_of_star_in_neighbourhood)
            #print(match.group(0), [line_id, match.span()[0], match.span()[1]-1])

print(numbers_with_star_neighbour[:5])
print(position_star_neighbour[:5])
```

```python
# "Super star" = exactly two numbers in neighbourhooud
position_super_star = [star_pos for star_pos in position_star_neighbour if position_star_neighbour.count(star_pos) == 2]
#pos_super_star
```

```python
unique_super_stars = []

for pos in position_super_star:
    if pos not in unique_super_stars:
        unique_super_stars.append(pos)
        #print(unique_super_stars)

print(unique_super_stars[:5])
```

```python
# Number with "super star" in neighbourhooud
products_per_super_star = []
temp_result = []
counter_super_stars_number = 0

for super_star in unique_super_stars:
    #print(super_star)
    res = 1
    num_ls = []
    for num, star in zip(numbers_with_star_neighbour, position_star_neighbour):
        if star == super_star:
            #print(num)
            num_ls.append(num)
            res *= int(num)
            counter_super_stars_number += 1
    #print(res)
    temp_result.append([super_star, num_ls])
    products_per_super_star.append(res)
    
    
#products_per_super_star
```

```python
result = sum(products_per_super_star)
print(f"Answer of puzzle 3 part 2 is {result}!")
```
