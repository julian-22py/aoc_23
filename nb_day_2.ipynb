# Advent of Code 2023 - Day 2

## Part 1

```python
import pandas as pd
import re
```

```python
with open('input/input_2.txt') as f:
    lines = f.readlines()
```

```python
valid_game_ids = []

for line in lines:
    #print(line)
    game_id = int(re.findall('Game (\d+):', line)[0])
    #print(game_id)
    checks = []
    # print(game_id, ':', line)
    for color, max_num in zip(['red', 'green', 'blue'],[12,13,14]):
        all_amounts = (re.findall(f'([0-9]+)(?:\s{color})', line))
        max_amount = max([int(amount) for amount in all_amounts])
        # print(color, max_amount)
        res = max_amount <= max_num
        checks.append(res)

    # print(checks) 
    if all(checks):
        valid_game_ids.append((game_id))

sum_valid_game_ids = sum(valid_game_ids)

print(f"The answer to part 1 is {sum_valid_game_ids}")
```

## Part 2

```python
products = []

for line in lines:
    #print(line)
    game_id = int(re.findall('Game (\d+):', line)[0])
    #print(game_id)
    minimums = []
    # print(game_id, ':', line)
    for color in ['red', 'green', 'blue']:
        all_amounts = (re.findall(f'([0-9]+)(?:\s{color})', line))
        max_color = max([int(amount) for amount in all_amounts])
        # print(color, max_amount)
        # res = max_amount <= max_num
        minimums.append(max_color)
    print(minimums)
    # print(checks)
    product = 1
    for i in minimums:
        product *= i
    print(product)
    products.append((product))

sum_products = sum(products)

print(f"The answer to part 2 is {sum_products}")
```
