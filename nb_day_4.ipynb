# Advent of Code 2023 - Day 3

```python
with open('input/input_4.txt') as f:
    games = f.read().splitlines()
```

```python
import re
```

Part 1

```python
def get_numbers_as_list(string):
    matches = re.finditer(r'\d+', string)
    numbers = []
    for num in matches:
        numbers.append(int(num.group(0)))
    return numbers

def get_common_elements(list1, list2, drop_duplicates = True):

    common_elements = [element for element in list1 if element in list2]

    if drop_duplicates:
        unique_elements = []

        for i in common_elements:
            if i not in unique_elements:
                unique_elements.append(i)
        res = unique_elements

    else:
        res = common_elements

    return res

# def unique_list(list):
#     unique_elements = []

#     for i in list:
#         if i not in unique_elements:
#             unique_elements.append(i)
#     return unique_elements

def get_points(game):
    winning_numbers = game[0]
    my_numbers = game[1]
    winning_numbers_match = get_common_elements(winning_numbers, my_numbers, False)

    length = len(winning_numbers_match)
    if length > 0:
        points = 2**(length-1)
    else:
        points = 0
    return points, winning_numbers_match

```

```python
points_per_game = []
winning_numbers = []

for game_id, game in enumerate(games):
    #print(game_id)
    game_without_card_id = game.split(':')[1]
    win, num = game_without_card_id.split('|')
    win_numbers = get_numbers_as_list(win)
    my_numbers = get_numbers_as_list(num)
    points, win_numbers = get_points([win_numbers, my_numbers])
    #print(points, win_numbers)

    points_per_game.append(points)
    winning_numbers.append(win_numbers)

#print(points_per_game)
```

```python
for id, numbers in enumerate(winning_numbers):
    #print((numbers))
    if any([numbers.count(i) > 1 for i in numbers]):
        print(id, numbers)
```

```python
result = sum(points_per_game)
print(f"Answer of puzzle 4 part 1 is {result}!")
```

Part 2

```python
num_copies = {i: 1 for i in range(len(winning_numbers))}
copies = 0

for i, numbers in enumerate(winning_numbers):
    n_win_numbers = len(numbers)
    id_copied_cards = [i + n + 1 for n in range(n_win_numbers)] * num_copies[i]
    #print(id_copied_cards)
    for j in id_copied_cards:
        num_copies[j] += 1

num_copies
```

```python
result = sum(num_copies.values())
print(f"Answer of puzzle 4 part 1 is {result}!")
```
